```python
# Parameters for our graph; we'll output images in a 4x4 configuration
nrows = 4
ncols = 4

# Index for iterating over images
pic_index = 0
```

```python
# Set up matplotlib fig, and size it to fit 4x4 pics
# Just two painters for now, Cezanne and Degas

fig = plt.gcf()
fig.set_size_inches(ncols * 4, nrows * 4)

pic_index += 8

next_cezanne_pix = [os.path.join(train_cezanne_dir, fname) 
                for fname in train_cezanne_names[pic_index-8:pic_index]]

next_degas_pix = [os.path.join(train_degas_dir, fname) 
                for fname in train_degas_names[pic_index-8:pic_index]]

for i, img_path in enumerate(next_cezanne_pix + next_degas_pix):
  # Set up subplot; subplot indices start at 1
  sp = plt.subplot(nrows, ncols, i + 1)
  sp.axis('Off') # Don't show axes (or gridlines)

  img = mpimg.imread(img_path)
  plt.imshow(img)

plt.show()
```
